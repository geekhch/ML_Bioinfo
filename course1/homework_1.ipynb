```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
iris = pd.read_csv('../iris_byWPS.csv')
```

```python
iris[:10]
```

```python
sl = 'sepal_length'
sw = 'sepal_width'
pl = 'petal_length'
pw = 'petal_width'
c = 'Class'
```

```python
c1 = iris[iris[c] == 'Iris-setosa']
c2 = iris[iris[c] == 'Iris-virginica']
c3 = iris[iris[c] == 'Iris-versicolor']
```

## 画图

```python
ax = plt.subplot(111, projection='3d')
ax.scatter(c1[pl], c1[pw], c1[sw],marker='2',label='se')
ax.scatter(c2[pl], c2[pw], c2[sw],marker='1',label = 'vi')
ax.scatter(c3[pl], c3[pw], c3[sw],marker='3', label = 've')
ax.legend()
#plt.legend((type1, type2, type3), ('不喜欢', '魅力一般', '极具魅力'))
```
